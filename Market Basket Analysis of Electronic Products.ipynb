```python
def encode(item_freq):
    res = 0
    if item_freq > 0:
        res = 1
    return res
    
basket_input = df2.applymap(encode)
```

```python
basket_input
```

```python
from mlxtend.frequent_patterns import apriori

itemsets = apriori(basket_input, min_support=0.001, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(itemsets, metric="lift", min_threshold=1)
```

```python
grouped = df.groupby('Order ID')['Product'].apply(list).reset_index()

expanded = pd.DataFrame(grouped['Product'].to_list(), index=grouped['Order ID'])

final_df = expanded.reset_index()
```

```python
final_df = final_df.drop('Order ID', axis =1)
```

```python
final_df
```

```python
from pyECLAT import ECLAT
eclat_instance = ECLAT(data=final_df, verbose=True)
```

```python
eclat_instance.df_bin  
eclat_instance.uniq_ 
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_instance.fit(min_support=0.08,
                                                           min_combination=1,
                                                           max_combination=3,
                                                           separator=' & ',
                                                           verbose=True)
```

```python
import pandas as pd
result = pd.DataFrame(get_ECLAT_supports.items(),columns=['Item', 'Support'])
result.sort_values(by=['Support'], ascending=False)
```
